# Metrics 
Monitoring how your queries perform is important. Your queries might not change but your data probably does. For example, a dimensions table could grow to be 10x what you assumed when creating the query.


## Monitor Response Time
We usually create endpoints for different use cases and track the response time continuously. If you want to go in depth to understand how your users are impacted by your response times you should use a histogram with your percentiles and calculate the likelihood of one user being affected by a slow query.

percentiles.jpg

Its important to just not look at the average time or 95% percentile specially in analytics contexts.

For example, if we have a dashboard that launches 10 requests at the same time and the average user sees 10 pages of that dashboard, if your 99% percentile is 10 seconds the 63.4% of the users will experience that 10 seconds request. Thats a lot, thats why max response times are important.

Where does this number come from? By definition, the probability of value being lower than the percentile X is X. Considering those 100 requests as independent events, the chances that all requests take less than the percentile 99 are 0.99 ^ 100 = 0.366. So the probability that one or more requests take more than that is 1 - 0.99 ^ 100 = 0.634 = 63.4%


## Benchmarking
Benchmarking is the best way to lie, the reason is that there are so many variables involved that it is easy to introduce bias by checking just a set of them.

Try to be honest with yourself and use a real load wherever  possible. Most of the time you dont know what the real load will be, so create a realistic simulation with estimations. Important things to take into account are:

 - QPS. Number of queries per second
 - Concurrency. How many clients will there be at the same time? A single client could generate several concurrent queries because dashboards usually have several widgets.
 - The probability distributions of filters and other parameters. Typically, users filter date columns by ranges of 1 day, 7 days, 1 month and so on. The probability of filtering by 1 day is usually 3-5x greater than filtering by 7 days. Take that into account when creating the benchmark.
